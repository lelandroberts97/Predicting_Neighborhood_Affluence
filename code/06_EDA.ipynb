# Exploratory Data Analysis

```python
# Imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading in the data
avg_price = pd.read_csv('../data/combined_3.csv')
price_category = pd.read_csv('../data/combined_4.csv')
```

## IRS Data

```python
# Reading in the data
irs = pd.read_csv('../data/combined_1.csv')
```

```python
sns.set(font_scale=2)
plt.figure(figsize=(15,8))
p = sns.distplot(irs['y'], color='coral', kde=False).set(xlabel='Price Bin',
                                                         ylabel='Count', 
                                                         title='Distribution of Income Bin');
```

## Average Price Data

```python
avg_price.head()
```

```python
# Heatmap to check correlations
plt.figure(figsize=(15,7))
sns.heatmap(avg_price.corr(), annot=True, cmap='coolwarm');
```

```python
# Pairplot to see relationships between variables
sns.pairplot(avg_price);
```

```python
# Checking the relationship between the highest correlated features
plt.scatter(data=avg_price, x='price', y='per_capita_income')
plt.xlabel('Average Restaurant Price')
plt.ylabel('Per Capita Income')
plt.title('Average Price vs Per Capita Income');
```

```python
plt.figure(figsize = (30, 30));
# columns = ['median_house_income', 'mean_house_income', 'per_capita_income']
fig, ax = plt.subplots(1,3, figsize=[15,6], sharex=True)

# plt.subplot(1,1,1)
sns.distplot(avg_price['median_household_income'], ax = ax[0]).set(xlabel='Median Household Income',
                                                ylabel='Count', 
                                                title='CT Median Household Income');

# plt.subplot(2,1,2)
sns.distplot(avg_price['mean_household_income'], ax = ax[1]).set(xlabel='Mean Household Income',
                                              ylabel='Count',
                                              title='CT Mean Household Income');

# plt.subplot(3,1,3)
sns.distplot(avg_price['per_capita_income'], ax = ax[2]).set(xlabel='Per Capita Income',
                                              ylabel='Count',
                                              title='CT Per Capita Income');
```

```python
sns.set(font_scale=2)
plt.figure(figsize=(15,8))
sns.distplot(avg_price['mean_household_income'], color='coral').set(xlabel='Mean Household Income',
                                                                    ylabel='Count',
                                                                    title='CT Mean Household Income');
```

```python
sns.set(font_scale=2)
plt.figure(figsize=(15,8))
sns.distplot(avg_price['mean_bucket'], kde=False, color='coral').set(xlabel='Mean Household Income',
                                                                     ylabel='Count',
                                                                     title='CT Mean Household Income');
```
