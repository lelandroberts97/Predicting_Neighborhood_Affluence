```python
# This function will do the process above for each zip code in a list of zip codes we input
def yelp_request(key, zip_codes, term = 'restaurants', limit =50):
    
    # Defining the endpoint and header
    yelp_url = 'https://api.yelp.com/v3/businesses/search'
    header = header = {'Authorization': f'bearer {key}'}
    
    # Defining an empty dictionary to store our restaurant data in
    r_dict = {}
    
    # Defining an empty data frame to append to
    df = pd.DataFrame(columns = ['id', 'name', 'review_count', 'rating', 'city', 'zip_code', 'state', 'price'])
    
    # Looping through each zip code to run a request for each
    for i in range(len(zip_codes)):
        
        # Defining our parameters
        params = {'term': 'restaurants',
          'limit': limit,
          'location': zip_codes[i]
        }
        
        # Making the request
        response = requests.get(url = yelp_url, params = params, headers = header)
        
        # Checking the status code
        #if response.status_code != 200:
            #raise
        
        # Converting JSON to a dictionary
        temp_dict = response.json()
        temp_dict = temp_dict['businesses']
        
        # Parsing the data 
        new_dict = {}
        for i in range(len(temp_dict)):
            try: 
                new_dict[i] = {
                    'id': temp_dict[i]['id'],
                    'name': temp_dict[i]['name'],
                    'review_count': temp_dict[i]['review_count'],
                    'rating': temp_dict[i]['rating'],
                    'city': temp_dict[i]['location']['city'],
                    'zip_code': temp_dict[i]['location']['zip_code'],
                    'state': temp_dict[i]['location']['state'],   
                    'price': temp_dict[i]['price']
            } 
            except: continue

        # Creating a data frame
        temp_df = pd.DataFrame(new_dict)
        temp_df = temp_df.T

        # Adding to complete data frame
        df = df.append(temp_df)
        
        # Waiting a second before scraping again
        time.sleep(1)
    
    # Saving to a csv
    df.to_csv('../data/yelp_2.csv', index=False)
    
    # Returning the final data frame
    return df
```
